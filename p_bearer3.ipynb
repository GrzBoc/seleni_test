```python
# Import the required modules
from selenium import webdriver
from selenium.webdriver.common.desired_capabilities import DesiredCapabilities
import time
import json
  
# Main Function
if __name__ == "__main__":
  
    # Enable Performance Logging of Chrome.
#     desired_capabilities = DesiredCapabilities.CHROME
#     desired_capabilities["goog:loggingPrefs"] = {"performance": "ALL"}

    # Create the webdriver object and pass the arguments
    options = webdriver.ChromeOptions()
  
    # Chrome will start in Headless mode
    options.add_argument('headless')
  
    # Ignores any certificate errors if there is any
    options.add_argument("--ignore-certificate-errors")

    options.set_capability("goog:loggingPrefs", {'performance': 'ALL'})      
    # Startup the chrome webdriver with executable path and
    # pass the chrome options and desired capabilities as
    # parameters.
    driver = webdriver.Chrome(options =options,
                              #desired_capabilities=desired_capabilities
                             )
  
    # Send a request to the website and let it load
    driver.get("https://www.geeksforgeeks.org/")
  
    # Sleeps for 10 seconds
    time.sleep(10)
  
    # Gets all the logs from performance in Chrome
    logs = driver.get_log("performance")
  
    # Opens a writable JSON file and writes the logs in it
    with open("network_log.json", "w", encoding="utf-8") as f:
        f.write("[")
  
        # Iterates every logs and parses it using JSON
        for log in logs:
            network_log = json.loads(log["message"])["message"]
  
            # Checks if the current 'method' key has any
            # Network related value.
            if("Network.response" in network_log["method"]
                    or "Network.request" in network_log["method"]
                    or "Network.webSocket" in network_log["method"]):
  
                # Writes the network log to a JSON file by
                # converting the dictionary to a JSON string
                # using json.dumps().
                f.write(json.dumps(network_log)+",")
        f.write("{}]")
  
    print("Quitting Selenium WebDriver")
    driver.quit()
  
    # Read the JSON File and parse it using
    # json.loads() to find the urls containing images.
    json_file_path = "network_log.json"
    with open(json_file_path, "r", encoding="utf-8") as f:
        logs = json.loads(f.read())
  
    # Iterate the logs
    for log in logs:
  
        # Except block will be accessed if any of the
        # following keys are missing.
        try:
            # URL is present inside the following keys
            url = log["params"]["request"]["url"]
  
            # Checks if the extension is .png or .jpg
            if url[len(url)-4:] == ".png" or url[len(url)-4:] == ".jpg":
                print(url, end='\n\n')
        except Exception as e:
            pass
```

```python
from selenium import webdriver
from selenium.webdriver.common.desired_capabilities import DesiredCapabilities
import time
import json
  
# Main Function
if __name__ == "__main__":
  
    # Enable Performance Logging of Chrome.
#     desired_capabilities = DesiredCapabilities.CHROME
#     desired_capabilities["goog:loggingPrefs"] = {"performance": "ALL"}

    # Create the webdriver object and pass the arguments
    options = webdriver.ChromeOptions()
  
    # Chrome will start in Headless mode
    options.add_argument('headless')
  
    # Ignores any certificate errors if there is any
    options.add_argument("--ignore-certificate-errors")

    options.set_capability("goog:loggingPrefs", {'performance': 'ALL'})      
    # Startup the chrome webdriver with executable path and
    # pass the chrome options and desired capabilities as
    # parameters.
    driver = webdriver.Chrome(options =options,
                              #desired_capabilities=desired_capabilities
                             )
  
    # Send a request to the website and let it load
    driver.get("https://www.geeksforgeeks.org/")
    
    # Inject JavaScript code to intercept XHR requests
    script = """
    (function() {
        var open = XMLHttpRequest.prototype.open;
        XMLHttpRequest.prototype.open = function() {
            this._url = arguments[1];
            return open.apply(this, arguments);
        };
        var send = XMLHttpRequest.prototype.send;
        XMLHttpRequest.prototype.send = function() {
            this.addEventListener('load', function() {
                var xhrLog = {
                    method: 'Network.request',
                    params: {
                        request: {
                            url: this._url,
                            method: arguments[0],
                            postData: arguments[0] === 'POST' ? JSON.stringify(arguments[1]) : undefined
                        }
                    }
                };
                console.log(JSON.stringify(xhrLog));
            });
            return send.apply(this, arguments);
        };
    })();
    """  
    driver.execute_script(script)
    
    # Sleeps for 10 seconds
    time.sleep(10)
  
    # Gets all the logs from performance in Chrome
    logs = driver.get_log("performance")

    logs = driver.get_log("browser")

    # List to store XHR requests
    xhr_requests = []

    # Iterate over the logs and parse the captured XHR requests
    for log in logs:
        log_entry = json.loads(log["message"])

        # Check if the log entry corresponds to an XHR request
        if log_entry["method"] == "Network.request":
            xhr_requests.append(log_entry)

    # Save the captured XHR requests
    with open("xhr_requests.json", "w") as xhr_file:
        json.dump(xhr_requests, xhr_file, indent=2)    
    
    
    
    
#      # Lists to store XHR, document, and request payload data
#     xhr_requests = []
#     document_requests = []
#     request_payloads = []

#     # Iterate over the logs and parse them using JSON
#     for log in logs:
#         network_log = json.loads(log["message"])["message"]

#         # Check if the current 'method' key has any Network related value.
#         method = network_log["method"]
#         print(network_log)
#         if ("Network.response" in method or "Network.request" in method or "Network.webSocket" in method):
#             # Check if it's an XHR request
#             if "Network.request" in method and 'request' in network_log["params"] and "xhr" in network_log["params"]["request"]["url"]:
#                 xhr_requests.append(network_log)
#                 # Extract request payload data if available
#                 if "postData" in network_log["params"]["request"]:
#                     request_payloads.append(network_log["params"]["request"]["postData"])

#             # Check if it's a document request
#             if "Network.request" in method and 'request' in network_log["params"] and "document" in network_log["params"]["request"]["url"]:
#                 document_requests.append(network_log)

#     # Save the captured XHR requests, document requests, and request payload data
#     with open("xhr_requests.json", "w") as xhr_file:
#         json.dump(xhr_requests, xhr_file, indent=2)

#     with open("document_requests.json", "w") as doc_file:
#         json.dump(document_requests, doc_file, indent=2)

#     with open("request_payloads.txt", "w") as payload_file:
#         payload_file.write("\n".join(request_payloads))

    # Quit the driver and close the browser
    driver.quit()
```

```python
from seleniumwire import webdriver  # Import from seleniumwire
from seleniumwire.utils import decode
import pandas as pd

# Ignores any certificate errors if there is any
options.add_argument("--ignore-certificate-errors")

options.set_capability("goog:loggingPrefs", {'performance': 'ALL'})      
# Startup the chrome webdriver with executable path and
# pass the chrome options and desired capabilities as
# parameters.

 # Ask the server not to compress the response
options_sele = {'disable_encoding': True }


driver = webdriver.Chrome(options =options,
                          seleniumwire_options=options_sele
                          #desired_capabilities=desired_capabilities
                         )
# Create a new instance of the Chrome driver
driver = webdriver.Chrome()

 


# Go to the Google home page
driver.get('https://www.google.com')
time.sleep(10)

logR=pd.DataFrame(columns=[ 'method','response','date','host','url','path','body','body_decode','cert','headers','params','querystring'])
logRes=pd.DataFrame(columns=[ 'reason','status_code','date','headers','body' ])

# Access requests via the `requests` attribute
all_requests=driver.requests
for request in all_requests:
    tmp_logR=pd.DataFrame(columns=[ 'method','date','host','url','path','body','body_decode','cert','headers','params','response','querystring'])
    tmp_logRes=pd.DataFrame(columns=[ 'status_code','date','headers','body','body_decode','reason'])
    if request.response:
        tmp_logR.loc[0,'method']=request.method
        tmp_logR.loc[0,'date']=request.date
        tmp_logR.loc[0,'host']=request.host
        tmp_logR.loc[0,'url']=request.url
        tmp_logR.loc[0,'path']=request.path
        tmp_logR.loc[0,'body']=request.body
        tmp_logR.loc[0,'body_decode']=request.body.decode('utf-8')
        tmp_logR.loc[0,'cert']=str(request.cert)
        tmp_logR.loc[0,'headers']=str(request.headers)
        tmp_logR.loc[0,'params']=str(request.params)
        tmp_logR.loc[0,'querystring']=request.querystring
        tmp_logR.loc[0,'response']=request.response

        #https://stackoverflow.com/questions/74969671/seleniumwire-get-response-text
        tmp_logRes.loc[0,'reason']=request.response.reason
        tmp_logRes.loc[0,'status_code']=request.response.status_code    
        tmp_logRes.loc[0,'date']=request.response.date    
        tmp_logRes.loc[0,'headers']=str(request.response.headers)
        tmp_logRes.loc[0,'body']=request.response.body   
        #tmp_logRes.loc[0,'body_decode']=request.response.body.decode('utf-8')  
        tmp_logRes.loc[0,'body_decode']=decode(request.response.body, request.response.headers.get('Content-Encoding', 'identity'))
        logR = pd.concat([logR, tmp_logR])
        logRes = pd.concat([logRes, tmp_logRes])        
           
driver.quit()

print('_______DONE________')
```
